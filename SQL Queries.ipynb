**2.Explain the difference between CROSS JOIN and INNER JOIN.**

1.**CROSS JOIN** returns the Cartesian product of two tables
   * Every row from table A is combined with
   * Every row from table B
   No condition is required

SELECT e.name, d.department_name
FROM employees e
CROSS JOIN departments d;

2**INNER JOIN**

1.An **INNER JOIN** returns only the rows match a specified condition

you must provide a join condition(usually using ON)

Depends on how many rows satisfy the join condition.

SELECT e.name, d.department_name
FROM employees e
INNER JOIN departments d
ON e.department_id = d.department_id;

**INNER JOIN** = match rows based on condition

**CROSS JOIN** = combine every row with every row

A CROSS JOIN can generate very large result sets quickly.

Example:

10,000 rows × 10,000 rows = 100 million rows

This can:

Consume memory

Slow queries

Overload the system





**3.What is query optimization?**

1.SQL optimization means improve SQL query performance to select most effective execution plan for execution of the SQL query.so it runs faster and uses fewer system resources.

When you write a SQL query, the database does not execute it exactly as written. Instead, the query optimizer analyzes it and chooses the best execution plan.

**Query optimization ensures:**

   1.Faster response times

   2.Lower CPU usage

   3.Reduced memory consumption

   4.Efficient disk I/O

**⚙️ How Query Optimization Works**

**When you submit a query:**

Parsing – The SQL syntax is validated.

Logical Plan Creation – The database figures out what needs to be done.

Cost Estimation – Different execution strategies are evaluated.

Execution Plan Selection – The lowest-cost plan is chosen.

Execution – The query runs using that plan.

**Common Optimization Techniques**

**✅ Indexing**

Create indexes on frequently filtered or joined columns.

**✅ Avoid SELECT **

Retrieve only required columns.

**✅ Proper Join Order**

Smaller tables should typically be joined first.

**✅ Use WHERE Filters Early**

Reduce rows as soon as possible.

**✅ Update Statistics**

Outdated statistics lead to bad execution plans.

**🚨 Signs a Query Needs Optimization**

Slow execution time

High CPU usage

High disk reads

Full table scans on large tables

Inefficient join operations


**🔄 Types of Optimization**

  **1️⃣ Rule-Based Optimization**

   Uses predefined rules

   Older approach

   Less flexible

2**️⃣ Cost-Based Optimization (Modern Systems)**

  Estimates cost (CPU, I/O, memory)

  Chooses lowest-cost plan

  Used in PostgreSQL, MySQL, SQL Server, Oracle, etc.
